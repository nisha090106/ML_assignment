Import Libraries and Create Dataset

```python
import pandas as pd
data = {
    "Student_ID": [101, 102, 103, 104],
    "Name": ["Nisha", "Priyanka", "Rahul", "Sneha"],
    "Age": [20, 21, 19, 22],
    "Gender": ["Female", "Female", "Male", "Female"],
    "Department": ["CSE", "ECE", "IT", "ME"],
    "Marks": [85, 92, 76, 88],
    "Attendance": [90, 95, 85, 80],
    "Internship": ["Yes", "No", "Yes", "No"]
}

df = pd.DataFrame(data)

print(df)
```

Shape of DataFrame

```python
print(df.shape)
```

Missing Values in Each Column

```python
print(df.isnull().sum())
```

Data Types of Each Column

```python
print(df.dtypes)
```

Zero Values Count

```python
print((df == 0).sum())
```

Fetch Columns

```python
print("Selecting first two rows:")
print(df.iloc[:2])

print("Selecting Name & Marks Columns:")
print(df[['Name', 'Marks']])
```

Sorted by Marks

```python
print(df.sort_values(by="Marks"))
```

 Describe Attributes of Data

```python
print(df.describe(include='all'))
```

Data Types Rechecked

```python
print(df.info())
```

Counting Unique Values

```python
print(df.nunique())
```

Column Names

```python
print(df.columns)
```

Converting Data Types

```python
df['Student_ID'] = df['Student_ID'].astype('int16')
df['Age'] = df['Age'].astype('int8')
df['Marks'] = df['Marks'].astype('int8')

print("Data Types After Converting:")
print(df.dtypes)
```

3.	Visualize the data using Python by plotting the graphs for assignment no. 1 and 2. Consider a suitable data set. Use Scatter plot, Bar plot, Box plot, Pie chart, Line Chart.


```python
import pandas as pd

# Create DataFrame
data = {
    'Student': ['A', 'B', 'C', 'D', 'E'],
    'Maths': [78, 85, 62, 90, 70],
    'Science': [72, 88, 65, 92, 75],
    'English': [68, 80, 70, 85, 73]
}

df = pd.DataFrame(data)
print(df)
```

Scatter Plot

Purpose: Relationship between two numerical variables (Maths vs Science)

```python
import matplotlib.pyplot as plt

plt.scatter(df['Maths'], df['Science'])
plt.xlabel('Maths Marks')
plt.ylabel('Science Marks')
plt.title('Scatter Plot: Maths vs Science')
plt.show()
```

Bar Plot

Purpose: Compare marks of students in Maths

```python
plt.bar(df['Student'], df['Maths'])
plt.xlabel('Students')
plt.ylabel('Maths Marks')
plt.title('Bar Plot: Maths Marks of Students')
plt.show()
```

Box Plot

Purpose: Show distribution and outliers of marks

```python
plt.boxplot([df['Maths'], df['Science'], df['English']],
            labels=['Maths', 'Science', 'English'])
plt.title('Box Plot of Subject Marks')
plt.ylabel('Marks')
plt.show()
```

Pie Chart

Purpose: Percentage contribution of total marks by subject (average)

```python
average_marks = df[['Maths', 'Science', 'English']].mean()

plt.pie(average_marks, labels=average_marks.index, autopct='%1.1f%%')
plt.title('Pie Chart: Average Marks Distribution')
plt.show()
```

Line Chart

Purpose: Trend of marks across students

```python
plt.plot(df['Student'], df['Maths'], label='Maths')
plt.plot(df['Student'], df['Science'], label='Science')
plt.plot(df['Student'], df['English'], label='English')

plt.xlabel('Students')
plt.ylabel('Marks')
plt.title('Line Chart: Marks Trend')
plt.legend()
plt.show()
```
